```python
# --- Import Library ---
from google_play_scraper import reviews_all
import pandas as pd
```

```python
# --- Scraping Review ---
app_id = 'com.kopikenangan'

reviews = reviews_all(
    app_id,
    lang='id',       # Bahasa Indonesia
    country='id'     # Negara Indonesia
)
```

```python
# --- Konversi ke DataFrame ---
df = pd.DataFrame(reviews)[['userName', 'content', 'score']]

# --- Urutkan dari Terbaru ---
df = df.iloc[::-1].reset_index(drop=True)

df = df.head(10000)

print("Total ulasan:", len(df))
```

```python
# --- Simpan ke CSV ---
df.to_csv('kopken_reviews.csv', index=False)
print("Data berhasil disimpan ke kopken_reviews.csv")

# --- Lihat Data ---
df.head()
```
